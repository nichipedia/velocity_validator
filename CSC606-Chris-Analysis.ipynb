```python
import pandas as pd
import numpy as np
```

```python
# Load a CSV file named 'data.csv' located in the same directory as the script
df = pd.read_csv('c:/csc606/ForNick/Data/PWAVE_LOGGER.csv',low_memory=False)

# Convert velocity column to numeric
df['compressional_velocity(m/s)'] = pd.to_numeric(df['compressional_velocity(m/s)'], errors='coerce')
```

```python
df.dropna(subset=['depth(mbsf)','compressional_velocity(m/s)'], inplace=True)
##a,b = np.polyfit( df["depth(mbsf)"], df["compressional_velocity(m/s)"], 1)
##df['distances'] = np.abs(a * df["depth(mbsf)"] - df["compressional_velocity(m/s)"] + b) / np.sqrt(a**2 + 1)
##df['r2'] = np.corrcoef(df['compressional_velocity(m/s)'], a*df['depth(mbsf)'])[0,1] ** 2
```

```python
groupby_columns = ['leg','site','hole','core','section']
df_grouped = df.groupby(groupby_columns)[['depth(mbsf)','compressional_velocity(m/s)']].agg(
    depth_mean=('depth(mbsf)','mean'),
    depth_median=('depth(mbsf)','median'),
    depth_mode=('depth(mbsf)',lambda x: x.mode()[0]),
    depth_std=('depth(mbsf)', 'std'),
    depth_min=('depth(mbsf)', 'min'),
    depth_max=('depth(mbsf)', 'max'),
    velocity_mean=('compressional_velocity(m/s)','mean'),
    velocity_median=('compressional_velocity(m/s)','median'),
    velocity_mode=('compressional_velocity(m/s)',lambda x: x.mode()[0]),
    velocity_std=('compressional_velocity(m/s)', 'std'),
    velocity_min=('compressional_velocity(m/s)', 'min'),
    velocity_max=('compressional_velocity(m/s)', 'max'),
    velocity_q1 =('compressional_velocity(m/s)',lambda x: x.quantile(0.25)),
    velocity_q3 =('compressional_velocity(m/s)',lambda x: x.quantile(0.75))
)
```

```python
df_grouped['velocity_igr'] = df_grouped['velocity_q3']-df_grouped['velocity_q1']
df_grouped['velocity_upper'] = df_grouped['velocity_median'] + (df_grouped['velocity_igr']*1.5)
df_grouped['velocity_lower'] = df_grouped['velocity_median'] - (df_grouped['velocity_igr']*1.5)
```
